# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('summer.csv')
```

```python
df
```

### 1. In how many cities Summer Olympics is held so far?

```python
print ("Summer Olympics has been held in " + str(df['City'].nunique()) + " cities so far.")
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
sport_gold = df[df.Medal=='Gold'].Sport.value_counts().head()
print(sport_gold)
sport_gold.plot(kind = 'bar', figsize = (10,10))
plt.title("Sports with most Gold Medals",)
plt.xlabel("Sport")
plt.ylabel("No. of Gold Medals")
plt.show()
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
sport_medal = df.groupby('Sport')['Medal'].count().sort_values(ascending = False).head()
print (sport_medal)
sport_medal.plot(kind = 'bar', figsize = (10,10))
plt.title("Sports with most Medals",)
plt.xlabel("Sport")
plt.ylabel("No. of Medals")
plt.show()
```

### 4. Which player has won most number of medals? (Top 5)

```python
plyr_medal=df.Athlete.value_counts().head()
print (plyr_medal)
plyr_medal.plot(kind = 'bar', figsize = (10,10))
plt.title("Athletes with most Medals",)
plt.xlabel("Athletes")
plt.ylabel("No. of Medals")
plt.show()
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
plyr_gold = df[df.Medal == 'Gold'].Athlete.value_counts().head()
print(plyr_gold)
plyr_gold.plot(kind = 'bar', figsize = (10,10))
plt.title("Athletes with most Gold Medals",)
plt.xlabel("Athlete")
plt.ylabel("No. of Gold Medals")
plt.show()
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
ind = df[df.Medal=='Gold']
print ('India won its first gold medal in Summer Olympics in the year '+ str(min(ind[ind.Country == 'IND'].Year))+'.')
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
popular_event = df['Event'].value_counts().head()
popular_event.plot(kind = 'bar', figsize = (10,10))
print(popular_event)
plt.title("Popular Event",)
plt.xlabel("Event")
plt.ylabel("No. of Players")
plt.show()
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
fml = df[df.Medal == 'Gold']
fml_gold = fml[fml.Gender == 'Women'].Sport.value_counts().head()
fml_gold.plot(kind = 'bar', figsize = (10,10))
print(fml_gold)
plt.title("Sport with more Female Gold Medalists",)
plt.xlabel("Sport")
plt.ylabel("No. of Female Gold Medalists")
plt.show()
```
